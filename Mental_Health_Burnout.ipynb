```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Mental_Health_Burnout.csv')
print(df.head())
```

```python
df.shape
```

```python
df.drop(columns = ['social_support_score'], inplace = True)
```

```python
df.columns
```

```python
df['seeks_mental_health_support']=df['seeks_mental_health_support'].map({1: 'yes', 0: 'No'})
print(df)
```

```python
df['job_change_intention']=df['job_change_intention'].map({1: 'yes', 0: 'No'})
print(df)
```

```python
df.isnull().sum()
```

```python
print('Job_satisfaction_score: ',df['job_satisfaction_score'].mean())
print('Stress_score: ',df['stress_score'].mean())
print('Burnout_score: ',df['burnout_score'].mean())
print('Manager_support_score: ',df['manager_support_score'].mean())
print('Burnout_level: ',df['burnout_level'].mode()[0])
```
